```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

### 1st method - sklearn

```python
x = np.array([49, 69, 89, 99, 109])
y = np.array([124, 95, 71, 45, 18])

X = np.vstack([np.ones([len(x)]), x]).T

print(X)

reg = LinearRegression().fit(X, y);

R_squared = reg.score(X, y)
coef = reg.coef_[1] 
intercept = reg.intercept_

plt.plot(x, y, 'ko')
plt.plot(x, coef*x+intercept, ":")
plt.xlabel('input')
plt.ylabel('output')
plt.text(80, 80, "$R^2 = {:.3f}$".format(R_squared))

print('The slope was calculated to be', coef)
print('The intercept was calculated to be', intercept)
```

### 2nd method - scipy

```python
L = linregress(x,y)
print(L)
print('......')
print('y = {}x + {}'.format(L.slope, L.intercept))
print('......')
print('The correlation coefficient was calculated to be {}'.format((L.rvalue)**2))
```
