```python
from scipy.optimize import newton
import numpy as np
import numdifftools as nd
import matplotlib.pyplot as plt

f = lambda x: -x**4 + 3*x**2

x = np.linspace(-2, 2)

# Find the root (solution)
initial_guess = np.linspace(-10, 10)
solution = np.round(newton(lambda x: nd.Derivative(f)(x), initial_guess), 3)

roots = np.unique(solution)
print('The roots are', roots)

plt.axhline(0, color='black', linewidth=0.5)  # Horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5)  # Vertical line at x=0
plt.plot(x, nd.Derivative(f)(x), label = "f'(x)")
plt.plot(x, f(x), label = 'f(x)')
plt.scatter(roots, np.zeros(len(roots)))
plt.legend()
```
