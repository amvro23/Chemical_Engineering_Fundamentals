# Shooting Method

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
```

$$ \frac{d^2y}{dt^2} = -g $$

with the two boundary conditions are: \\( y(0) = 0 \\) and \\(y(5) = 50 \\). We would like to fing the \\( y~'(0) \\)

```python
g = 9.8
t = np.linspace(0, 5, 100)
y0 = 0
v0 = 25

def f_ODE(t, u):
    return np.array([u[1], -g])

u0 = [y0, v0]

def objective(x):
    # where x is a vector that contains the unknows initial values
    sol = solve_ivp(f_ODE, [t[0], t[-1]], [0, x[0]])
    y = sol.y[0]
    return y[-1] - 50

v0, = fsolve(objective, 10)
print(v0)

# update the initial values
u0 = [y0, v0]

sol_ODE = solve_ivp(f_ODE, [t[0], t[-1]], u0, t_eval = t)

plt.plot(sol_ODE.t, sol_ODE.y[0], 'k')
plt.plot(5, 50, 'yo')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.title('root finding v={:.1f} m/s'.format(v0))
plt.show()
```

Testing different initial guesses

```python
for v0_guess in range(1, 100, 10):
    v0, = fsolve(objective, v0_guess)
    print('Init: %d, Result: %.1f' \
          %(v0_guess, v0))
```

### Using solve_bvp

```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt

g = 9.8
y0 = 0
v0 = 25

def f_ode(t, u):
    size = len(u[0])
    return np.vstack((u[1], -g*np.ones(size)))

def bc(ya, yb):
    left = ya[0]         # for y'(0) it would have been ya[1]
    right = yb[0]-50     # for y'(5) it would have been yb[1]
    return np.array([left, right])

t = np.linspace(0, 5, 100)

y_0 = np.zeros((2, t.size))

res = solve_bvp(f_ode, bc, t, y_0)

plt.plot(res.x, res.y[0], 'g')
plt.plot(5, 50, 'ms')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.title('root finding v={:.1f} m/s'.format(v0))
plt.show()
```
