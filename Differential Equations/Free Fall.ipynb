# Free fall

\\( \ddot{y} = -g\\) or

\\( \frac{\mathrm{d}^2y}{\mathrm{d}t^2} = f(t,y,\dot{y}) = -g\\)

Analytical solution: \\( y(t)=-\frac{g}{2}t^2+v_0t+y_0\\)

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

g = 9.81
v0 = 50
y0 = 0

def f_ODE(t, y):
    return np.array([y[1], -g])
```

We will need two initial values for \\( y, \dot{y} \\), respectively. 
- Note that \\(y \\) represents the distance & \\(\dot{y} \\) represents the velocity
- \\(\ddot{y} \\) represents the acceleration.

```python
# initial values for y, 
init = np.array([y0, v0])

t = np.linspace(0.0, 10.0, 100)

sol = solve_ivp(f_ODE, (0, t[-1]), init, t_eval = t)

distance = sol.y[0]
velocity = sol.y[1]
```

```python
max_y_index = np.argmax(sol.y[0])
mat_t_index = t[max_y_index]
```

```python
y_analytical = -g/2*t**2 + v0*t + y0

plt.plot(t, y_analytical, 'k.', label = 'analytical solution', lw = 5)
plt.plot(t, distance, color = 'lime', ls = '--', label = 'numerical RK45')
plt.plot(t, velocity, color = 'm', ls = '-', label = 'velocity')
plt.xlabel('t (time)')
plt.ylabel('x (distance)')

plt.axhline(0, linestyle = '--', alpha = 0.3)
plt.axvline(mat_t_index, linestyle = '--', alpha = 0.3)
plt.axhline(-g, color = 'y', linestyle = '-', alpha = 0.3, label = 'g')


plt.legend()
```

### Observations

We can see that the velocity is zero when we have the maximum distance:

- \\( \frac{\mathrm{d}y}{\mathrm{d}t} = 0\\)

### Testing a different initial value

```python
y02 = 30

# initial values for y, 
init2 = np.array([y02, v0])

t = np.linspace(0.0, 10.0, 100)

sol2 = solve_ivp(f_ODE, (0, t[-1]), init2, t_eval = t)

distance2 = sol2.y[0]
velocity2 = sol2.y[1]

y_analytical2 = -g/2*t**2 + v0*t + y02

plt.plot(t, y_analytical, 'k.', label = 'analytical solution', lw = 5)
plt.plot(t, distance, color = 'lime', ls = '--', label = 'numerical RK45')
plt.plot(t, velocity, color = 'm', ls = '-', label = 'velocity')
plt.xlabel('t (time)')
plt.ylabel('x (distance)')

plt.plot(t, y_analytical2, 'k.', label = 'analytical solution ', lw = 5)
plt.plot(t, distance2, color = 'lime', ls = '--', label = 'numerical RK45')
plt.plot(t, velocity2, color = 'm', ls = '-', label = 'velocity')


plt.legend()
```

*Here we can see the effect of changing the initial value without changing the derivative.*
