```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
```

```python
x = np.array([0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7, 1.9, 2])
y = np.array([0.57,0.85,0.66,0.84,0.59,0.55,0.61,0.76,0.54,0.55,0.48])

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, y, color = "grey", marker = "o", linestyle = '-')
ax.set_ylim(0.5, 1.0)
ax.grid(linestyle='dotted')

# smoothed line
x_new = np.linspace(x.min(), x.max(), 35)
f = interp1d(x, y, kind='quadratic')
y_smooth = f(x_new)

# [x0, y0, width, height]
axIns = ax.inset_axes([0.55, 0.55, 0.4, 0.4])

axIns.plot(x_new, y_smooth, "g-", label = "smoothed")
axIns.plot(x, y, "go")
axIns.set_ylim(0.5, 1.0)
axIns.grid(linestyle='dotted')
axIns.legend(fontsize = 8)
```
