```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

***(a) Grouped Bar Chart***

```python
data = {'material_1': [94.0, 48.0, 74.0],
        'material_2': [75.0, 73.0, 102.0],
        'material_3': [79.0, 73.0, 65.0],
        'material_4': [121.0, 55.0, 72.0]}

labels = ['25 Degrees', '35 Degrees','45 Degrees']

df = pd.DataFrame(data, index = labels)
df
```

```python
# the label locations
x = np.arange(len(labels))  

# the width of the bars
width = 0.20
```

```python
# plot the bars
fig, ax = plt.subplots(figsize=(6,4), dpi=100)

rects1 = ax.bar(x - width/2, df.material_1, width, color='k', edgecolor='w', hatch = '/////', label='material_1')
rects2 = ax.bar(x + width/2, df.material_2, width, color='deepskyblue', edgecolor='k', hatch = '/////', label='material_2')
rects3 = ax.bar(x - 3*width/2, df.material_3, width, color='deeppink', edgecolor='k', hatch = '/////', label='material_3')
rects4 = ax.bar(x + 3*width/2, df.material_4, width, color='gold', edgecolor='k', hatch = '/////', label='material_4')

# add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('y axis', fontsize=12, fontweight='bold')
ax.set_xlabel('x axis', fontsize=12, fontweight='bold')
plt.xticks(x, labels)


ax.bar_label(rects1, padding=1, fontsize=7.0)
ax.bar_label(rects2, padding=1, fontsize=7.0)
ax.bar_label(rects3, padding=1, fontsize=7.0)
ax.bar_label(rects4, padding=1, fontsize=7.0)


for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
    ax.spines[axis].set_color('k')
    ax.spines[axis].set_zorder(1)
    
plt.xticks(weight = 'bold')
plt.yticks(weight = 'bold')

plt.text(-1.15, 125.0, '(a)', fontsize=20, fontweight='bold')

params = {'legend.fontsize': 8, 'font.weight': 'bold'}
plt.rcParams.update(params)
ax.legend(loc='upper center', frameon=False)
plt.title('Grouped Bar Chart', loc='center', fontsize=12, fontweight='bold')
```

***(b) Single Bar Chart***

```python
data = {'mat1':190.0 , 'mat2':266.0, 'mat3':138.0, 'mat4':183.0}

materials = list(data.keys())
values = list(data.values())
```

```python
# creating the bar plot
fig, ax = plt.subplots(figsize=(6,4), dpi=100)
    
ax.bar(materials, values, color = ['deeppink', 'k', 'deepskyblue', 'gold'],
        width = 0.6, edgecolor = ['k', 'w', 'k', 'k'], 
        hatch = ['....', '....', '....', '....'])

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
    ax.spines[axis].set_color('k')
    ax.spines[axis].set_zorder(1)

plt.xticks(weight = 'bold')
plt.yticks(weight = 'bold')    

# to display values above the bar
xlocs, xlabs = plt.xticks()
for i, v in enumerate(values):
    ax.text(xlocs[i] - 0.15, v + 2.0, str(v), fontsize=7, fontweight='bold')


ax.set_ylabel("y axis", fontsize=12, fontweight='bold')
ax.set_xlabel("x axis", fontsize=12, fontweight='bold')
ax.set_title("Single Bar Chart", fontsize=12, fontweight='bold', loc='center')
ax.text(-1.15, 270.0, '(b)', fontsize=20, fontweight='bold')
```
